```python
# De 4 bevegelseslikningene med konstant akselerasjon

def bevegelseslikninger(u, a, t):
    # u: initial hastighet (m/s)
    # a: akselerasjon (m/s^2)
    # t: tid (s)
    
    # Likning 1: v = u + at
    v = u + a * t
    
    # Likning 2: s = ut + 0.5 * a * t^2
    s = u * t + 0.5 * a * t**2
    
    # Likning 3: v^2 = u^2 + 2as
    v_squared = u**2 + 2 * a * s
    
    # Likning 4: s = ((u + v) / 2) * t
    s_avg = ((u + v) / 2) * t
    
    return v, s, v_squared, s_avg

# Eksempel på bruk
u = 10  # initial hastighet i m/s
a = 2   # akselerasjon i m/s^2
t = 5   # tid i sekunder

v, s, v_squared, s_avg = bevegelseslikninger(u, a, t)

print(f"Slutt hastighet (v): {v} m/s")
print(f"Tilbakelagt distanse (s): {s} m")
print(f"Slutt hastighet kvadrert (v^2): {v_squared} m^2/s^2")
print(f"Tilbakelagt distanse (gjennomsnittlig hastighet) (s_avg): {s_avg} m")
```

```python
# Newtons første lov: Et legeme forblir i ro eller i jevn bevegelse med konstant hastighet hvis ingen ytre krefter virker på det.
def newtons_first_law(force):
    if force == 0:
        return "Legemet forblir i ro eller i jevn bevegelse."
    else:
        return "Legemet akselererer."

# Newtons andre lov: F = ma (kraft = masse * akselerasjon)
def newtons_second_law(mass, acceleration):
    force = mass * acceleration
    return force

# Newtons tredje lov: For hver kraft er det en like stor og motsatt rettet kraft.
def newtons_third_law(force):
    reaction_force = -force
    return reaction_force

# Eksempel på bruk
force = 0  # ingen ytre kraft
print(newtons_first_law(force))

mass = 5  # masse i kg
acceleration = 2  # akselerasjon i m/s^2
print(f"Kraft (F): {newtons_second_law(mass, acceleration)} N")

force = 10  # kraft i N
print(f"Reaksjonskraft: {newtons_third_law(force)} N")
```

```python
# Eksempel 7-7 En krets med en lampe, batteri og en motstand er koplet i serie
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# Gitte verdier
R_lamp = 80  # Resistansen til lampa i [ohm]
R_resistor = 200  # Resistansen til motstanden i [ohm]
V_battery = 4.0  # Spenningen til batteriet i [volt]

# Beregn total resistans i kretsen
R_total = R_lamp + R_resistor

# Beregn strømmen ved hjelp av Ohms lov
I = V_battery / R_total

# Beregn spenningen over lampa ved hjelp av Ohms lov
V_lamp = I * R_lamp

# Beregn spenningen over motstanden ved hjelp av Ohms lov
V_resistor = I * R_resistor

# Print resultatene
print(f"Den totale resistansen i kretsen er: {R_total} Ohm")
print(f"Strømmen i kretsen er: {I:.3f} Ampere")
print(f"Spenningen over lampa blir: {V_lamp:.1f} Volt")
print(f"Spenningen over motstanden blir: {V_resistor:.1f} Volt")

# Sjekk om beregningene stemmer overens med Kirchhoffs 2. lov
print("Summen av spenningene over komponentene:", (V_lamp + V_resistor), "Volt er det samme som spenningen på batteriet", (V_battery), "i Volt")
if abs(V_battery - (V_lamp + V_resistor)) < 1e-6:
    print("Beregningene stemmer overens med Kirchhoffs 2. lov.")
else:
    print("Beregningene stemmer ikke overens med Kirchhoffs 2. lov.")
```
